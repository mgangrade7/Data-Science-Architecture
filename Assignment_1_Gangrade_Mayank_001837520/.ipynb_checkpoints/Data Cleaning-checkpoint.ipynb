### Checking Null Values on all the columns
There are various method to check null values througout the columns:

* .apply() : Applies function along input axis of DataFrame
* .isnull() : Detect missing values (NaN in numeric arrays, None/NaN in object arrays)


```python
# Method : 1
wur_data.apply(lambda x: sum(x.isnull()),axis=0) # Apply : Applies function along input axis of DataFrame
```

```python
# Method : 2
wur_data.isnull().sum()
```
