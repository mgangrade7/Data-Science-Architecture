### Data Set
#### World University Rankings
#### Data Set Source : https://www.kaggle.com/mylesoneill/world-university-rankings/data

```python
# Importing CSV in to python
import pandas as pd

data = pd.read_csv("DataSet.csv", sep=',')
```

```python
data.head()

# Column Details:
# world_rank  world rank for university Numeric
# institution name of university String
# country country of each university String
# national_rank rank of university within its country Numeric
# quality_of_education rank for quality of education Numeric
# alumni_employment rank for alumni employment Numeric
# quality_of_faculty rank for quality of faculty Numeric
# publications rank for publications Numeric
# influence rank for influence Numeric
# citations rank for citations Numeric
# broad_impact rank for broad impact(only available for 2014 and 2015) String
# patents rank for patents Numeric
# score total score, used for determining world rank Numeric
# year year of ranking (2012 to 2015) Numeric
```

### Cleaning and EDA

```python
# Importing Pakages
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
from sklearn.datasets import load_boston

# Make plots larger
plt.rcParams['figure.figsize'] = (15, 9)
```

```python
data.head()
```

```python
# Checking for missing/null values
data.isnull().sum()
```

```python
# Checking ration of null vs not null broad_impact
null_broad_impact = sum(data['broad_impact'].isnull())
total_broad_impact = len(data['broad_impact'])
ratio = (null_broad_impact / total_broad_impact) * 100
print(null_broad_impact,total_broad_impact)
print(ratio)
```

```python
# Replacing values : only 9% values are missing for 'broad_impact' so i am replacing these values with 0
data['broad_impact'].fillna(0, inplace=True)
data.isnull().sum()
```

```python
# Checking for inappropraite values
data.info()
```

```python
# Data backup
bak_data = data.copy()
```

```python
# As i am not going to do any analysis on alumni_employment,influence and citations so i am deleting these columns

del bak_data['alumni_employment']
del bak_data['influence']
del bak_data['citations']
bak_data.info()
```

```python
# Rename: quality_of_faculty to faculty_rank
bak_data.rename(columns={'quality_of_faculty' : 'faculty_rank'}, inplace=True)
bak_data.info()
```

```python
# summary statistics of all column
bak_data.describe()
```

```python
bak_data.shape
```

```python
# Checking for Corelation between the columns
bak_data.corr()
# Conclusion
# 1. A perfect uphill (positive) linear relationshipn between world_rank and publications, world_ranh and broad_impact
# 2. A strong uphill (positive) linear relationship between quality_of_education and faculty_rank
# 3. A perfect uphill (positive) linear relationshipn between publications and world_rank
# 4. A moderate downhill (negative) linear relationship between world_rank and score
# 5. A moderate downhill (negative) relationship between score and faculty_rank
```

```python
# Analysis : Number of universities with respect to country
id_count_by_region = bak_data.groupby('country')['world_rank'].count()
id_count_by_region.sort_values(na_position='last', inplace=True, ascending=False)
id_count_by_region[:25]
```

```python
# Analysis : Finding number of institutes with respect to the country
dt_country = bak_data.country.value_counts()

fig = plt.figure(figsize=(6, 14))
dt_country.plot(kind='barh')
plt.title('Total Institute by Country')
plt.xlabel('Counts')
plt.show()
```
