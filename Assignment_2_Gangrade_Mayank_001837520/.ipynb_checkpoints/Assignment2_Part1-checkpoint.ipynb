```python
%matplotlib inline
from collections import Counter
import random

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN
from scipy.stats import multivariate_normal
```

```python
data_frame = pd.read_csv("DataSet.csv")
print(data_frame.shape)
data_frame.head()

# Column Details:
# world_rank  world rank for university Numeric
# institution name of university String
# country country of each university String
# national_rank rank of university within its country Numeric
# quality_of_education rank for quality of education Numeric
# alumni_employment rank for alumni employment Numeric
# quality_of_faculty rank for quality of faculty Numeric
# publications rank for publications Numeric
# influence rank for influence Numeric
# citations rank for citations Numeric
# broad_impact rank for broad impact(only available for 2014 and 2015) String
# patents rank for patents Numeric
# score total score, used for determining world rank Numeric
# year year of ranking (2012 to 2015) Numeric
```

#### Clustering
Methods used for clutering
1. k-means clustering
2. DBSCAN

```python
# k-means clustering
# Getting the values and plotting it
f1 = data_frame['national_rank'].values
f2 = data_frame['quality_of_education'].values
X = np.array(list(zip(f1, f2)))
plt.scatter(f1, f2, c='black', s=10)
```

```python
# Fit a k-means estimator
estimator = KMeans(n_clusters=8)
X = data_frame[["national_rank", "quality_of_education"]]
estimator.fit(X)
# Clusters are given in the labels_ attribute
labels = estimator.labels_
print(labels)
```

```python
def set_colors(labels, colors='rgbykcmw'):
    colored_labels = []
    for label in labels:
        if (label < 0) or (label > 6):
          colored_labels.append(colors[7]) 
        else:
          colored_labels.append(colors[label])
    return colored_labels

colors = set_colors(labels)
plt.scatter(data_frame['national_rank'], data_frame['quality_of_education'], c=colors)
plt.title("KMeans = 2")
plt.xlabel("national_rank")
plt.ylabel("quality_of_education")
plt.show()
```

```python
estimator = KMeans(n_clusters= 2)
X = data_frame[["national_rank", "quality_of_education"]]
estimator.fit(X)
# Clusters are given in the labels_ attribute
labels = estimator.labels_
print(Counter(labels))

colors = set_colors(labels)
plt.scatter(data_frame['national_rank'], data_frame['quality_of_education'], c=colors)
plt.title("KMeans k=2")
plt.xlabel("national_rank")
plt.ylabel("quality_of_education")
plt.show()
```

```python
# DBSCAN 
# Fit a DBSCAN estimator
estimator = DBSCAN(eps=3, min_samples=10)
X = data_frame[["national_rank", "quality_of_education"]]
estimator.fit(X)
# Clusters are given in the labels_ attribute
labels = estimator.labels_
# print labels
print(Counter(labels))

colors = set_colors(labels)
plt.scatter(data_frame['national_rank'], data_frame['quality_of_education'], c=colors)
plt.xlabel("national_rank")
plt.ylabel("quality_of_education")
plt.show()
```

##### Answer the following questions for the clustering:

* Do the clustering methods generate the same clusters?
> Yes, clusters are same in both (DBSCAN and k-means) clustering methods.

* Does scaling effect the clustering? 
> Yes, In DBSCAN scaling effect the clustering, As we increasing the value of eps(The maximum distance between two samples for them to be considered as in the same neighborhood) the output started covering more and more points.

* Does the clustering produce interesting groupings?
> Yes, with the change in the value of k cluster started becomes more and more specific.
